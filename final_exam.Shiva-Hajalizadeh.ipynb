```python
import matplotlib.pyplot as plt

def plot_scores(students):
    """
    رسم نمودار نمرات درس‌های سیستم مخابراتی و محاسبات عددی
    - رنگ آبی برای سیستم مخابراتی
    - رنگ قرمز برای محاسبات عددی
    - نمایش میانگین هر درس در نمودار با همان رنگ
    """
    names = [s["first_name"] for s in students]
    communication_scores = [s["scores"]["communication_system"] for s in students]
    numerical_scores = [s["scores"]["numerical_calculations"] for s in students]

    plt.plot(names, communication_scores, marker="o", linestyle="--", color="blue", label="Communication System")
    plt.plot(names, numerical_scores, marker="s", linestyle="-", color="red", label="Numerical Calculations")

    plt.axhline(sum(communication_scores)/len(communication_scores), color="blue", linestyle="dashed", label="Avg Communication")
    plt.axhline(sum(numerical_scores)/len(numerical_scores), color="red", linestyle="dashed", label="Avg Numerical")

    avg_communication = sum(communication_scores) / len(communication_scores)

    
    plt.legend()
    plt.title("Scores of Students")
    plt.xlabel("Students")
    plt.ylabel("Scores")
    plt.show()

# داده‌های نمرات
students = [
    {"first_name": "arvin", "scores": {"communication_system": 8, "numerical_calculations": 18}},
    {"first_name": "shiva", "scores": {"communication_system": 10, "numerical_calculations": 16}},
    {"first_name": "parsa", "scores": {"communication_system": 12, "numerical_calculations": 14}},
    {"first_name": "mahdi", "scores": {"communication_system": 14, "numerical_calculations": 12}},
    {"first_name": "fateme", "scores": {"communication_system": 16, "numerical_calculations": 10}},
    {"first_name": "hasan", "scores": {"communication_system": 18, "numerical_calculations": 8}},
]

plot_scores(students)
```

```python
def categorize_scores(students):
    """
    تبدیل نمرات به یکی از چهار دسته A, B, C, D
    A: نمرات بین 17 تا 20
    B: نمرات بین 14 تا 17
    C: نمرات بین 10 تا 14
    D: نمرات بین 0 تا 10
    """
    categories = {"A": 0, "B": 0, "C": 0, "D": 0}

    for s in students:
        for score in s["scores"].values():
            if 17 <= score < 20:
                categories["A"] += 1
            elif 14 < score <= 17:
                categories["B"] += 1
            elif 10 < score <= 14:
                categories["C"] += 1
            elif 0 <= score <= 10:
                categories["D"] += 1
    
    return categories

def plot_pie_chart(categories):
    """
    رسم نمودار دایره‌ای بر اساس توزیع نمرات در دسته‌های A, B, C, D
    - رنگ طلایی برای گروه A
    - رنگ آبی روشن برای گروه B
    - رنگ سالمونی برای گروه C
    - رنگ خاکستری برای گروه D
    """
    labels = categories.keys()
    sizes = categories.values()
    colors = ["gold", "lightblue", "salmon", "gray"]

    plt.pie(sizes, labels=labels, autopct="%1.1f%%", colors=colors, startangle=90)
    plt.title("Grade Distribution")
    plt.show()

students = [
    {"first_name": "arvin", "scores": {"communication_system": 8, "numerical_calculations": 18, "linear_control": 17, "energy_system": 16}},
    {"first_name": "shiva", "scores": {"communication_system": 10, "numerical_calculations": 16, "linear_control": 16, "energy_system": 17}},
    {"first_name": "parsa", "scores": {"communication_system": 12, "numerical_calculations": 14, "linear_control": 15, "energy_system": 12}},
    {"first_name": "mahdi", "scores": {"communication_system": 14, "numerical_calculations": 12, "linear_control": 8, "energy_system": 13}},
    {"first_name": "fateme", "scores": {"communication_system": 16, "numerical_calculations": 10, "linear_control": 17, "energy_system": 8}},
    {"first_name": "hasan", "scores": {"communication_system": 18, "numerical_calculations": 8, "linear_control": 12, "energy_system": 16}},
]

categories = categorize_scores(students)
plot_pie_chart(categories)
```
